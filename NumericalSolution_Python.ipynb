### **Solution to in-class example**

Enter the constants of the mechanism in the cell below

```python
R1, R2, R3, R4, theta1, theta2 = 5, 3, 7, 6, 0, 60 * np.pi / 180
```

Provide your initial guesses in the cell below

First try $\theta_3=10$ degrees and $\theta_4=100$ degrees

```python
x = np.array([[10 * np.pi / 180], [100 * np.pi / 180]])
```

Run the cell below for Newton-Raphson iterations on this problem

```python
tolerance = 0.01 #adjust the tolerance value here

theta3, theta4 = x[0, 0], x[1, 0]
print(f"Initial Guess is:\n theta3={x[0, 0] * 180 / np.pi:.4f} degrees,\n theta4={x[1, 0] * 180 / np.pi:.4f} degrees\n")

i = 1

while True:
    # Non-linear functions column vector
    f = np.array([
        [R2 * np.cos(theta2) + R3 * np.cos(theta3) - R4 * np.cos(theta4) - 5],
        [R2 * np.sin(theta2) + R3 * np.sin(theta3) - R4 * np.sin(theta4)]
    ])

    # Jacobian matrix
    J = np.array([
        [-R3 * np.sin(theta3), R4 * np.sin(theta4)],
        [R3 * np.cos(theta3), -R4 * np.cos(theta4)]
    ])

    # Newton-Raphson update
    x = x - np.linalg.inv(J)@f

    print(f"Iteration number: {i}\n")
    print(f"Newton-Raphson Guess:\n theta3={x[0, 0] * 180 / np.pi:.4f} degrees,\n theta4={x[1, 0] * 180 / np.pi:.4f} degrees\n")

    # Convergence check
    if abs(theta3 * 180 / np.pi - x[0, 0] * 180 / np.pi) < tolerance and abs(theta4 * 180 / np.pi - x[1, 0] * 180 / np.pi) < tolerance:
        break

    # Update thetas for next iteration
    theta3, theta4 = x[0, 0], x[1, 0]

    i += 1

# Final solution
print(f"Newton-Raphson solutions:\n theta3={theta3 * 180 / np.pi:.4f} degrees, theta4={theta4 * 180 / np.pi:.4f} degrees")
```
